```python
cat_cols = ['description', 'engine', 'trim', 'exterior_color', 'interior_color']
for col in cat_cols:
    data[col] = data[col].fillna('Unknown')
```

```python
data['cylinders'] = data['cylinders'].fillna(data['cylinders'].mode()[0])
data['mileage'] = data['mileage'].fillna(data['mileage'].median())
data['doors'] = data['doors'].fillna(data['doors'].mode()[0])
```

```python
data['fuel'] = data['fuel'].fillna(data['fuel'].mode()[0])
data['body'] = data['body'].fillna(data['body'].mode()[0])
data['transmission'] = data['transmission'].fillna(data['transmission'].mode()[0])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# 🌟 Hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# 🔍 Grid Search with Cross-Validation
rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# 🏋️‍♀️ Fit time!
grid_search.fit(X_train, y_train)

# ✅ Best model found
print(f"Best Parameters: {grid_search.best_params_}")

# 🔮 Predict with best estimator
best_rf_model = grid_search.best_estimator_
y_pred = best_rf_model.predict(X_test)

# 📈 Evaluation again
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f"Tuned Mean Squared Error (Random Forest): {mse:.2f}")
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# 🌟 Randomized Search Param Grid
param_dist = {
    'n_estimators': randint(50, 300),
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None] + list(range(5, 30)),
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 10),
}

# 🌳 Random Forest Model
rf_model = RandomForestRegressor(random_state=42)

# 🏃 Randomized Search CV
random_search = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_dist,
    n_iter=150,  # You can bump this up for more thorough search
    cv=3,
    verbose=2,
    n_jobs=-1,
    scoring='neg_mean_squared_error',
    random_state=42
)

# 🔥 Fit dat model
random_search.fit(X_train, y_train)

# 🏆 Best Params & Score
print(f"Best Parameters (Randomized Search): {random_search.best_params_}")

# 📊 Predict & Evaluate
y_pred = random_search.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Tuned Mean Squared Error (Randomized Search RF): {mse:.2f}")
```

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error
import numpy as np

# Model instance
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

# Hyperparameter grid
param_dist = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [3, 5, 7, 10, None],
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'min_child_weight': [1, 3, 5, 7],
    'gamma': [0, 0.1, 0.2, 0.3]
}

# Randomized Search
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=100,  # You can play with this (50, 100, 150)
    cv=3,
    scoring='neg_mean_squared_error',
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# Fit it!
random_search.fit(X_train, y_train)

# Best params & predictions
print(f"Best Parameters (XGBoost Randomized Search): {random_search.best_params_}")

# Predictions & MSE
y_pred_xgb = random_search.predict(X_test)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
print(f"Tuned Mean Squared Error (XGBoost): {mse_xgb:.2f}")
```

```python
import matplotlib.pyplot as plt
xgb.plot_importance(random_search.best_estimator_, max_num_features=10, importance_type='gain')
plt.title('Top 10 Feature Importances (XGBoost)')
plt.show()
```

```python
# 🎯 Extract fitted model from RandomizedSearchCV
xgb_best_model = random_search.best_estimator_

# ✅ Now get feature importances
importances = xgb_best_model.feature_importances_

# 📊 Build DataFrame of importances
feature_importances_df = pd.DataFrame({
    'feature': X_train.columns,
    'importance': importances
})

# 🔥 Sort and pick top 20
top_features = feature_importances_df.sort_values(by='importance', ascending=False).head(20)['feature'].tolist()

# ✂️ Subset train & test
X_train_top = X_train[top_features]
X_test_top = X_test[top_features]

# 🚀 Retrain with top features (optional, but flex)
xgb_model_top = xgb.XGBRegressor(
    n_estimators=500,
    max_depth=10,
    learning_rate=0.1,
    subsample=0.6,
    colsample_bytree=1.0,
    min_child_weight=1,
    gamma=0,
    random_state=42
)

xgb_model_top.fit(X_train_top, y_train)

# 📈 Predict & evaluate
y_pred_top = xgb_model_top.predict(X_test_top)
mse_top = mean_squared_error(y_test, y_pred_top)

print(f"Tuned MSE with Top Features (XGBoost): {mse_top:.2f}")
```
